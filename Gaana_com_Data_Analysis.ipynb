```python
import pandas as pd
from tqdm import tqdm 

df=pd.read_csv('Gaana-Data-Analysis/songs.csv')
```

```python
df.isnull().sum()
```

2.Data Preprocessing


   Dealing with the duration




```python
duration_lst =[]
for duration in df["duration"]:

  Min=int(duration.split(':')[0])
  Sec=int(duration.split(':')[1])
  Sec=(Min*60)+Sec

  duration_lst.append(Sec)

df["duration"]=duration_lst

df.head()
```

```python
len(duration_lst)
```

```python
 print("Average duration of a song",int(df['duration'].sum()/len(df["duration"])))
```

```python
#Average duration of a song throughout the set
s=0
 
for i in df["duration"]:
  s+=i

print(s/len(df['duration']))
```

```python
#Average duration of a song in each lang
for lang in df["language"].unique():
  print(lang, int(df[df["language"]==lang].mean()))
```

```python
for lang in df["language"].unique():
  print(lang, len(df[df["language"]==lang]))
```

```python
#How many singers are there in the dataset?
all_singers=[]
for singers in df['singer']:
  all_singers += singers.split("|")

print("Total Singers :",len(set(all_singers)))
  
```

```python
#unique singers
unique_singers =list(set(all_singers))

for i in unique_singers:
  print(i)
```

```python
#checking no. of songs of jagjit singh
c=0
for i in df['singer']:
  if("Jagjit Singh"==i):
    c+=1
print(c)
```

```python
c=0
for i in df['singer']:
  if("Jagjit Singh"in i):
    c+=1
print(c)
```

```python
#Who has most no. of songs played?
data=[]
unique_singers =list(set(all_singers))

#tqdm tells how much time it will take for execution
for i in tqdm(unique_singers):
  c=0
  for singers in df['singer']:
    if(i in singers):
      c+=1
  data.append([i,c])
```

```python
songs = pd.DataFrame(data,columns=["singer","songs"])
songs.head()
```

```python
songs.sort_values(by="songs",ascending=False).head()
```
